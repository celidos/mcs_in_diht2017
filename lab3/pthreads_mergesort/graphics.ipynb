```python
import matplotlib.pyplot as plt
import numpy as np
```

Возьмем следующие параметры:

$P \in \{1,2,4,8,16\}$

$N = 10^7$&nbsp;&nbsp;&nbsp;$(\approx 10 \text{  МБ})$

$m = 50000$

`3775 3578 10000000 50000 1
2249 3676 10000000 50000 2
1590 3609 10000000 50000 4
1632 3597 10000000 50000 8
1626 3581 10000000 50000 16`

```python
P = np.array([1, 2, 4, 8, 16])

time_ms = np.array([3775, 2249, 1590, 1632, 1626]) # merge sort
time_qs = np.array([3578, 3676, 3609, 3597, 3581]) # quick sort from library
```

**График зависимости $T(P)$:**

```python
plt.figure(figsize=(15, 9))
plt.xlabel('$P$')
plt.ylabel('$T$, мс')
plt.title('График зависимости времени работы от количества потоков')
plt.grid(ls=':')

plt.plot(P, time_ms, '.-', label='mergesort time')
plt.plot(P, time_qs, '.-', label='quicksort time', color='red')
plt.legend()
plt.show()
```

```python
S = time_ms[0] / time_ms
E = S / P
```

**График зависимости $S(P)$**:

```python
plt.figure(figsize=(14, 8))
plt.xlabel('P')
plt.ylabel('S')
plt.title('График зависимости ускорения S(P) от количества потоков P')
plt.grid(ls=':')

plt.plot(P, S, '.-', label='S')
plt.legend()
plt.show()
```

**График зависимости $E(P)$:**

```python
plt.figure(figsize=(14, 8))
plt.xlabel('P')
plt.ylabel('E')
plt.title('График зависимости эффективности E(P) от количества потоков P')
plt.grid(ls=':')

plt.plot(P, E, '.-', label='E')
plt.legend()
plt.show()
```
